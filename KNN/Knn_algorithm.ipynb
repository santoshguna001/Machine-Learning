```python
#importing necessary libraries for this problem
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
#seaborn is used to visualize data in an effective manner
```

```python
#Loading the data from the file and dividing them into test data, train data, input data and output data
data=pd.read_csv("Prostate_Cancer.csv")
X=data.iloc[:,2:]
y_train=data.iloc[:90,1].ravel()
x_train=X.iloc[0:90,:]
x_test=X.iloc[90:100,:]
y_test=data.iloc[90:100,1].ravel()
```

```python
#Visualizing the dataset for every parameter against every other parameter including itself
sns.pairplot(data.iloc[1:,:], hue='diagnosis_result')
```

```python
#Classifying the data using the KNN algorithm 
knn = KNeighborsClassifier()
knn.fit(x_train, y_train) 
KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',metric_params=None, n_jobs=1, n_neighbors=5, p=2, weights='uniform')
```

```python
print("Cancer type(Malignant or Benign) detected from the KNN algorithm:")
predicted=knn.predict(x_test)
print(predicted)
print("Actual cancer type:")
print(y_test)
count=0
for i in range(len(predicted)):
    if(y_test[i] == predicted[i]):
        count=count+1
print('Efficiency is',count/len(y_test)*100.0)
```
