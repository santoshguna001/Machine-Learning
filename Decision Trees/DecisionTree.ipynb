```python
#importing necessary libraries
import pandas as pd
from sklearn import tree
#graphviz is used to draw graphs for the decision tree
import graphviz
#accuracy_score is used to compare the predicted results with the actual results
from sklearn.metrics import accuracy_score
```

```python
#Loading the data from the file and dividing them into test data, train data, input data and output data
data=pd.read_csv('Iris.csv')
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
data1=pd.read_csv('test.csv')
X1=data1.iloc[:,:-1]
Y1=data1.iloc[:,-1].ravel()
#X and Y have the training data
#X1 and Y1 have the testing data
features=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalLengthCm']
target=['1','2','3']
#features and target are initialized for drawing the graph later
```

```python
#implementation of decision tree considering the entropy as the condition to split a node for the tree
model = tree.DecisionTreeClassifier(criterion='entropy') 
decision=model.fit(X, Y)
predicted= model.predict(X1)
print('Predicted values are (using entropy as the condition):')
print(predicted)
print('Actual values are:')
print(Y1)
print('Efficiency is',accuracy_score(predicted,Y1)*100)



dot_data = tree.export_graphviz(decision, out_file=None,
                         feature_names=features,
                         class_names=target,
                         special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

```python
#implementation of decision tree considering the gini as the condition to split a node for the tree
model1 = tree.DecisionTreeClassifier()#By default the criteria is taken as gini, there is no need to specify
decision1=model1.fit(X, Y)
predicted= model1.predict(X1)
print('Actual values are:')
print(Y1)
print('Predicted values are (using gini as the condition):')
print(predicted)
print('Efficiency is',accuracy_score(predicted,Y1)*100)



dot_data = tree.export_graphviz(decision1, out_file=None,
                         feature_names=features,
                         class_names=target,
                         special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```
