```python
#import necessary libraries for the program
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
from sklearn.cross_validation import train_test_split
from sklearn import svm
import seaborn as sns
```

```python
#load the data and split them into target values and feature variables
data_original=pd.read_csv("data.txt", delimiter=',')
data=data_original.iloc[:,1:]
X=data.iloc[:,1:]
Y=data.iloc[:,0].ravel()

#spilt the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=42)
Y_train=Y_train.ravel()
Y_test=Y_test.ravel()

#visualize the data
sns.pairplot(data.iloc[1:,:], hue='diagnosis_result')
```

```python
#model using SVM for the training data
classifier_model = svm.SVC(kernel='linear', C=100,gamma=1).fit(X_train, Y_train)
Y_predict=classifier_model.predict(X_test)

#compute the efficiency
count=0
for i in range(len(Y_test)):
    if(Y_test[i]==Y_predict[i]):
        count=count+1
        
print((count/len(Y_test))*100)
```

```python
#model using SVM for the training data with different value of C - the regularization parameter
classifier_model = svm.SVC(kernel='linear', C=10,gamma=1).fit(X_train, Y_train)
Y_predict=classifier_model.predict(X_test)

#compute the efficiency
count=0
for i in range(len(Y_test)):
    if(Y_test[i]==Y_predict[i]):
        count=count+1
        
print((count/len(Y_test))*100)
```

```python
#model using SVM for the training data with different value of gamma
classifier_model = svm.SVC(kernel='linear', C=10,gamma=0.1).fit(X_train, Y_train)
Y_predict=classifier_model.predict(X_test)

#compute the efficiency
count=0
for i in range(len(Y_test)):
    if(Y_test[i]==Y_predict[i]):
        count=count+1
        
print((count/len(Y_test))*100)
```
